```python
import pandas as pd
habitantes=pd.read_excel("C:/CursPython/Datos Accidente Original.xlsx",sheet_name="Datos")
habitantes
```

```python
habitantes["Num. Accidentes"]
```

```python
import matplotlib.pyplot as plt
```

```python
y=habitantes["Num. Accidentes"]
x=habitantes["Habitantes"]
plt.scatter(x,y,color="blue")
plt.title("Numero d'acciodents de la població")
plt.show;
```

```python
y=habitantes["Num. Accidentes"]
x=habitantes["Habitantes"]
plt.scatter(x,y,color="blue")
plt.title("Numero d'acciodents de la població")
plt.show;
import numpy as np
z=np.polyfit(x,y,1)
h=np.poly1d(z)(x)
plt.plot(x,h,"r--",lw=1)
plt.show;
```

```python
from sklearn.linear_model import LinearRegression
mlineal=LinearRegression()
mlineal.fit(x.values.reshape(-1,1),y)
a=mlineal.coef_
b=mlineal.intercept_
print("El valor de la pendent a es",a)
print("El valor de la intersecció b a es", round(b,4))
```

```python
Poblacio_Ficticia=np.array([95000])
prediccio=mlineal.predict(Poblacio_Ficticia.reshape(-1,1))
print("El numero previst d'accident per una poblacio es", Poblacio_Ficticia, "sera",prediccio, "accidentts")
```

```python
r2=mlineal.score(x.values.reshape(-1,1),y)
print("r2=",round(r2,2))
```
