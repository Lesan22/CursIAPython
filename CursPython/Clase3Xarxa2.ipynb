```python
imatges_entrenament=imatges_entrenament/255.0
imatges_test=imatges_test/255.0
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(6,5,i+1)
    plt.xticks([]) #elimina llegenda x
    plt.yticks([])
    plt.imshow(imatges_entrenament[i],cmap=plt.cm.binary)
    plt.xlabel(nomRoba[etiquetes_entrenament[i]])
plt.show()
    
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(10,activation='softmax')])


```

```python
#Compilar el model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Entrenar
model.fit(imatges_entrenament, etiquetes_entrenament,epochs=10)
```
